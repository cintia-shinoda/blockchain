```python
import math
import matplotlib.pyplot as plt
```

```python
hash_power_rede = 90
```

```python
hash_power_fraudador = 10
```

```python
p = hash_power_rede / (hash_power_rede + hash_power_fraudador)
p
```

```python
q = 1 - p
q
```

```python
z = 2
```

```python
l = z * (q/p)
l
```

```python
soma = 1
```

```python
def probabilidade_fraude(q, z):
    p = 1 - q
    l = z * (q/p)
    soma = 1

    for k in range(z + 1):
        poisson = math.exp(-l)
        for i in range(1, k+1):
            poisson *= l / i

        soma -= poisson * (1 - math.pow(q/p, z-k))

    return soma
```

```python
probabilidade_fraude(q=0.3, z=5)
```

```python
zs = [2,3,4,5]
qs = [0.10, 0.20, 0.30, 0.40, 0.50]

for q in qs:
    dominus = []
    functus = []
    for z in zs:
        qz = probabilidade_fraude(q=q, z=z)
        dominus.append(z)
        functus.append(qz)

    plt.plot(dominus, functus)
    plt.legend(qs)
```
